```python
import matplotlib.pyplot as plt

# Aggregate data by key and calculate mean popularity
key_popularity = st.groupby("key")["popularity"].mean().reset_index()

# Sort by key for better visualization
key_popularity = key_popularity.sort_values(by="key")

# Define colors
bar_color = 'skyblue'

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(key_popularity["key"], key_popularity["popularity"], color=bar_color)
plt.title('Mean Popularity by Key')
plt.xlabel('Key')
plt.ylabel('Mean Popularity')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.savefig("Mean_Popularity_by_Key.png")
plt.show()
```

```python
max_popularity = st.groupby('key')['popularity'].max().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='key', y='popularity', data=max_popularity)
plt.title('Maximum Popularity for Each Key')
plt.xlabel('Key')
plt.ylabel('Maximum Popularity')
plt.savefig("Maximum_Popularity_for_Each_Key.png")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='key', hue='mode', data=st)
plt.title('Count of Records for Each Key, Grouped by Mode')
plt.xlabel('Key')
plt.ylabel('Count of Records')
plt.savefig("RecordsCountforEachKey_GroupedbyMode.png")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Choose a specific artist
selected_artist = "His Hot Seven"

# Filter data for the selected artist
artist_data = st[st['artists'].str.contains(selected_artist)]

# Sort data by release date
artist_data = artist_data.sort_values(by='release_date')

# Create a line plot
plt.figure(figsize=(12, 6))
sns.lineplot(x='release_date', y='popularity', data=artist_data, marker='o', color='green')
plt.title(f'Popularity of {selected_artist}\'s Tracks Over Time')
plt.xlabel('Release Date')
plt.ylabel('Popularity')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()
plt.savefig("line_plot.png")

plt.show()
```

```python
# Choose two numerical columns for the scatter plot
x_column = 'energy'
y_column = 'valence'

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x=x_column, y=y_column, data=st, alpha=0.7, color='blue')
plt.title(f'Scatter Plot: {x_column} vs {y_column}')
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.savefig("Scatter_Plot_Energy_vs_Valence.png")

plt.show()
```

```python
# Scatter plot between 'danceability' and 'energy'
sns.scatterplot(x='danceability', y='energy', data=st)
plt.title('Scatter Plot: Danceability vs Energy')
plt.savefig("Scatter_Plot_Danceability_vs_Energy.png")

plt.show()
```

```python
# Creating a box plot for the 'key' column
plt.figure(figsize=(10, 5))
plt.boxplot(st['key'], showmeans=True)
plt.title("Box plot of the column 'key'")
plt.xlabel("key")
plt.ylabel("Values")
plt.savefig("Box_Plot_Key.png")

plt.show()
```
